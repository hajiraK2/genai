```python
import gensim.downloader as api
```

```python
model = api.load("word2vec-google-news-300")
word1 = "king"
word2 = "queen"
similarity = model.similarity(word1, word2)
print("Similarity between 'king' and 'queen':", similarity)
```

```python
result = model.most_similar(positive=["king", "woman"], negative=["man"], topn=1)
print("'King' - 'Man' + 'Woman' =", result[0][0], "Similarity:", result[0][1])
similar_words = model.most_similar("Paris", topn=5)
print("Words similar to Paris:")
for word, score in similar_words:
    print(word, score)
odd_word = model.doesnt_match(["breakfast", "lunch", "dinner", "banana"])
print("Odd one out:", odd_word)
```
